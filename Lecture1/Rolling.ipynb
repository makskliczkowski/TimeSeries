# Rolling and expanding

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/Users/mirek/Downloads/SBUX1.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
df['Close'].plot(figsize=(12,6)).autoscale(axis='x',tight=True)
```

```python
df.rolling(window=7).mean().head(10)
```

```python
df['Close'].plot(figsize=(12,6)).autoscale(axis='x',tight=True)
df.rolling(window=30).mean()['Close'].plot()
```

```python
df['Close'].expanding(min_periods=30).mean().plot(figsize=(12,6))
```
